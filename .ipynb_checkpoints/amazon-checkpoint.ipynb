```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Setup ChromeDriver with options
def setup_driver():
    chrome_options = Options()
    chrome_options.add_argument("--start-maximized")  # Open browser in full screen
    # chrome_options.add_argument("--headless")  # Uncomment for headless mode
    chrome_options.add_argument("--disable-gpu")  # Disable GPU acceleration
    chrome_options.add_argument("--no-sandbox")  # Disable sandboxing for Linux systems
    service = Service(ChromeDriverManager().install())
    return webdriver.Chrome(service=service, options=chrome_options)

# Function to extract links on the current page
def extract_links(driver):
    links = []
    try:
        all_links = driver.find_elements(By.TAG_NAME, "a")  # Find all anchor tags
        for link in all_links:
            href = link.get_attribute('href')
            if href:  # Ensure the link is valid
                links.append(href)
        print(f"Extracted {len(links)} links from the current page.")
    except Exception as e:
        print(f"Error extracting links: {e}")
    return links

# Function to click the "Next" button and navigate through pages
def click_next_button(driver):
    all_links = []  # To store links from all pages
    try:
        while True:
            # Extract links from the current page
            page_links = extract_links(driver)
            all_links.extend(page_links)

            # Wait for the "Next" button and click it
            next_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//a[contains(@class, "s-pagination-next")]'))
            )
            print("Clicking the 'Next' button...")
            next_button.click()

            # Wait for the next page to load
            time.sleep(2)  # Optional: Adjust based on page load time
    except Exception as e:
        print(f"Navigation ended: {e}")
    finally:
        print(f"Total links collected: {len(all_links)}")
    return all_links

# Main script
if __name__ == "__main__":
    try:
        driver = setup_driver()
        driver.get("https://www.amazon.in/s?k=laptop")  # Replace with your target URL
        time.sleep(5)  # Allow initial content to load

        # Call the function to start extracting links
        links = click_next_button(driver)
        print("Collected links:")
        print(links)
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    finally:
        driver.quit()  # Ensure the browser is closed
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Setup ChromeDriver with options
def setup_driver():
    chrome_options = Options()
    chrome_options.add_argument("--start-maximized")  # Open browser in full screen
    # chrome_options.add_argument("--headless")  # Uncomment for headless mode
    chrome_options.add_argument("--disable-gpu")  # Disable GPU acceleration
    chrome_options.add_argument("--no-sandbox")  # Disable sandboxing for Linux systems
    service = Service(ChromeDriverManager().install())
    return webdriver.Chrome(service=service, options=chrome_options)

# Function to extract links on the current page
def extract_links(driver):
    links = []
    try:
        all_links = driver.find_elements(By.TAG_NAME, "a")  # Find all anchor tags
        for link in all_links:
            href = link.get_attribute('href')
            if href and href.startswith("http"):  # Ensure the link is valid and starts with "http"
                links.append(href)
        print(f"Extracted {len(links)} links from the current page.")
    except Exception as e:
        print(f"Error extracting links: {e}")
    return links

# Function to click the "Next" button and navigate through pages
def click_next_button(driver):
    all_links = []  # To store links from all pages
    try:
        while True:
            # Extract links from the current page
            page_links = extract_links(driver)
            all_links.extend(page_links)

            # Wait for the "Next" button and check if it's clickable
            try:
                next_button = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, '//a[contains(@class, "s-pagination-next") and not(contains(@class, "s-pagination-disabled"))]'))
                )
                print("Clicking the 'Next' button...")
                next_button.click()

                # Wait for the page to load by waiting for the next button to become stale
                WebDriverWait(driver, 10).until(EC.staleness_of(next_button))

                # Optional wait for better accuracy
                time.sleep(2)
            except Exception as e:
                print("No more pages or unable to find 'Next' button. Ending pagination.")
                break
    except Exception as e:
        print(f"Error during pagination: {e}")
    finally:
        print(f"Total links collected: {len(all_links)}")
    return all_links

# Main script
if __name__ == "__main__":
    try:
        driver = setup_driver()
        driver.get("https://www.amazon.in/s?k=laptop")  # Replace with your target URL
        time.sleep(5)  # Allow initial content to load

        # Call the function to start extracting links
        links = click_next_button(driver)
        print("Collected links:")
        print(links)
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    finally:
        driver.quit()  # Ensure the browser is closed
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Setup ChromeDriver with options
def setup_driver():
    chrome_options = Options()
    chrome_options.add_argument("--start-maximized")  # Open browser in full screen
    # chrome_options.add_argument("--headless")  # Uncomment for headless mode
    chrome_options.add_argument("--disable-gpu")  # Disable GPU acceleration
    chrome_options.add_argument("--no-sandbox")  # Disable sandboxing for Linux systems
    service = Service(ChromeDriverManager().install())
    return webdriver.Chrome(service=service, options=chrome_options)

# Function to extract pagination links on the current page
def extract_pagination_links(driver):
    links = []
    try:
        # Find all pagination links (usually anchor tags in the pagination section)
        pagination_links = driver.find_elements(By.XPATH, '//a[contains(@class, "s-pagination-item")]')
        
        for link in pagination_links:
            href = link.get_attribute('href')
            if href:  # Ensure the link is valid
                links.append(href)
        print(f"Extracted {len(links)} pagination links from the current page.")
    except Exception as e:
        print(f"Error extracting pagination links: {e}")
    return links

# Function to click the "Next" button and navigate through pages, collecting pagination links
def click_next_button(driver):
    all_pagination_links = []  # To store pagination links from all pages
    try:
        while True:
            # Extract pagination links from the current page
            page_links = extract_pagination_links(driver)
            all_pagination_links.extend(page_links)

            # Wait for the "Next" button and check if it's clickable
            try:
                next_button = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, '//a[contains(@class, "s-pagination-next") and not(contains(@class, "s-pagination-disabled"))]'))
                )
                print("Clicking the 'Next' button...")
                next_button.click()

                # Wait for the page to load by waiting for the next button to become stale
                WebDriverWait(driver, 10).until(EC.staleness_of(next_button))

                # Optional wait for better accuracy
                time.sleep(2)
            except Exception as e:
                print("No more pages or unable to find 'Next' button. Ending pagination.")
                break
    except Exception as e:
        print(f"Error during pagination: {e}")
    finally:
        print(f"Total pagination links collected: {len(all_pagination_links)}")
    return all_pagination_links

# Main script
if __name__ == "__main__":
    try:
        driver = setup_driver()
        driver.get("https://www.amazon.in/s?k=laptop")  # Replace with your target URL
        time.sleep(5)  # Allow initial content to load

        # Call the function to start extracting pagination links
        pagination_links = click_next_button(driver)
        print("Collected pagination links:")
        print(pagination_links)
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    finally:
        driver.quit()  # Ensure the browser is closed
```
